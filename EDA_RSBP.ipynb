# EDA

```python
import pandas as pd

df = pd.read_csv('/content/used_cars.csv')
display(df.head())
```

## Brand

```python
brand_counts = df['brand'].value_counts()
display(brand_counts)
```

## Manufacture Year

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.histplot(df['model_year'], kde=True, bins=30)
plt.title('Distribution of Model Year')
plt.xlabel('Model Year')
plt.ylabel('Frequency')
plt.show()
```

## Milage

```python
# Normalisasi
print("Tipe data sebelum cleaning:", df['milage'].dtype)
print("Contoh data sebelum cleaning:", df['milage'].head(3).to_list())
print("-" * 30)
df['milage'] = df['milage'].str.replace(' mi.', '', regex=False)
df['milage'] = df['milage'].str.replace(',', '', regex=False)
df['milage'] = pd.to_numeric(df['milage'], errors='coerce')

#Visualisasi Graph
plt.figure(figsize=(12, 6))
sns.histplot(df['milage'], kde=True, bins=20000)
plt.title('Distribution of Mileage')
plt.xlabel('Mileage')
plt.ylabel('Frequency')
plt.show()
```

## Fuel Type

```python
fuel_type_counts_counts = df['fuel_type'].value_counts()
display(fuel_type_counts_counts)
```

```python
df.fillna('not defined', inplace=True)
display(df.head())
```

# DATA CLEANING


```python
missing_values_count = df.isnull().sum()
display(missing_values_count)
```

# EXPORT CLEAN CSV

```python
df.to_csv('used_car_clean.csv', index=False)
print("Data berhasil diekspor ke used_car_clean.csv")
```

# More EDA


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert price to numeric, handling potential errors
df['price'] = df['price'].replace({'\$': '', ',': ''}, regex=True)
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# Scatter plot: Model Year vs Price, colored by Accident
plt.figure(figsize=(14, 6))
sns.scatterplot(data=df, x='model_year', y='price', hue='accident', alpha=0.6)
plt.title('Relationship between Model Year, Accident, and Price')
plt.xlabel('Model Year')
plt.ylabel('Price')
plt.grid(True)
plt.show()

# Scatter plot: Milage vs Price, colored by Accident
plt.figure(figsize=(14, 6))
sns.scatterplot(data=df, x='milage', y='price', hue='accident', alpha=0.6)
plt.title('Relationship between Milage, Accident, and Price')
plt.xlabel('Milage')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```
